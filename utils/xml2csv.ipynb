```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns                #导入了seaborn，matplotlib的默认作图风格就会被覆盖成seaborn的格式
import xml.etree.cElementTree as et
```

```python
dfcols = ['id', 'method', 'path', 'query', 'label']

def get_value(node):
    return node.text if node is not None else None

def xml2df(xml_path, csv_path):
    df_xml = pd.DataFrame(columns=dfcols)
    df_xml.id = df_xml.id.astype(int)
    parsed_xml = et.parse(xml_path)
    for node in parsed_xml.getroot():
        id = int(node.attrib.get('id'))
        method = get_value(node.find('request').find('method'))
        path = get_value(node.find('request').find('path'))
        if method == 'GET':
            query = get_value(node.find('request').find('query'))
        else:
            query = get_value(node.find('request').find('body'))
        type = get_value(node.find('label').find('type'))
        attack = get_value(node.find('label').find('attack'))
        label = attack if attack is not None else type
        
        df_xml = df_xml.append(pd.Series([id, method, path, query, label], index=dfcols), ignore_index=True)
    df_xml.to_csv(csv_path)
    return df_xml
```

```python
anomalies1_xml = r'../data/torpeda/anomalies/allAnomalies1.xml'
anomalies2_xml = r'../data/torpeda/anomalies/allAnomalies2.xml'
attacks1_xml = r'../data/torpeda/attacks/allAttacks1.xml'
attacks2_xml = r'../data/torpeda/attacks/allAttacks2.xml'
attacks3_xml = r'../data/torpeda/attacks/allAttacks3.xml'
attacks4_xml = r'../data/torpeda/attacks/allAttacks4.xml'
attacks5_xml = r'../data/torpeda/attacks/allAttacks5.xml'
normals1_xml = r'../data/torpeda/normals/allNormals1.xml'

anomalies1_csv = r'../data/torpeda_csv/anomalies/allAnomalies1.csv'
anomalies2_csv = r'../data/torpeda_csv/anomalies/allAnomalies2.csv'
attacks1_csv = r'../data/torpeda_csv/attacks/allAttacks1.csv'
attacks2_csv = r'../data/torpeda_csv/attacks/allAttacks2.csv'
attacks3_csv = r'../data/torpeda_csv/attacks/allAttacks3.csv'
attacks4_csv = r'../data/torpeda_csv/attacks/allAttacks4.csv'
attacks5_csv = r'../data/torpeda_csv/attacks/allAttacks5.csv'
normals1_csv = r'../data/torpeda_csv/normals/allNormals1.csv'
```

```python
xml2df(anomalies1_xml, anomalies1_csv)
xml2df(anomalies2_xml, anomalies2_csv)
xml2df(attacks1_xml, attacks1_csv)
xml2df(attacks2_xml, attacks2_csv)
xml2df(attacks3_xml, attacks3_csv)
xml2df(attacks4_xml, attacks4_csv)
xml2df(attacks5_xml, attacks5_csv)
xml2df(normals1_xml, normals1_csv)
```
