```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns                #导入了seaborn，matplotlib的默认作图风格就会被覆盖成seaborn的格式
from pandas.core.frame import DataFrame
```

```python
anomalies1_csv = r'../data/torpeda_csv/anomalies/allAnomalies1.csv'
anomalies2_csv = r'../data/torpeda_csv/anomalies/allAnomalies2.csv'
attacks1_csv = r'../data/torpeda_csv/attacks/allAttacks1.csv'
attacks2_csv = r'../data/torpeda_csv/attacks/allAttacks2.csv'
attacks3_csv = r'../data/torpeda_csv/attacks/allAttacks3.csv'
attacks4_csv = r'../data/torpeda_csv/attacks/allAttacks4.csv'
attacks5_csv = r'../data/torpeda_csv/attacks/allAttacks5.csv'
normals1_csv = r'../data/torpeda_csv/normals/allNormals1.csv'
```

```python
anomalies1_df = pd.read_csv(anomalies1_csv)
anomalies2_df = pd.read_csv(anomalies2_csv)
attacks1_df = pd.read_csv(attacks1_csv)
attacks2_df = pd.read_csv(attacks2_csv)
attacks3_df = pd.read_csv(attacks3_csv)
attacks4_df = pd.read_csv(attacks4_csv)
attacks5_df = pd.read_csv(attacks5_csv)
normals1_df = pd.read_csv(normals1_csv)
```

```python
def get_attrib(s):
    rst = anomalies1_df[s].tolist()+anomalies2_df[s].tolist()+\
        attacks1_df[s].tolist()+attacks2_df[s].tolist()+attacks3_df[s].tolist()+\
        attacks4_df[s].tolist()+attacks5_df[s].tolist()+normals1_df[s].tolist()
    return rst
```

```python
def isNaN(query):
    return query != query

def get_url(method, path, query):
    url = []
    for m, p, q in zip(method, path, query):
        if isNaN(q):
            u = m + ' ' + p
        else:
            u = m + ' ' + p + '?' + q
        url.append(u)
    return url
```

```python
file = [anomalies1_csv for i in range(len(anomalies1_df))] + \
       [anomalies2_csv for i in range(len(anomalies2_df))] + \
       [attacks1_csv for i in range(len(attacks1_df))] + \
       [attacks2_csv for i in range(len(attacks2_df))] + \
       [attacks3_csv for i in range(len(attacks3_df))] + \
       [attacks4_csv for i in range(len(attacks4_df))] + \
       [attacks5_csv for i in range(len(attacks5_df))] + \
       [normals1_csv for i in range(len(normals1_df))]

id = get_attrib('id')
method = get_attrib('method')
path = get_attrib('path')
query = get_attrib('query')
label = get_attrib('label')
url = get_url(method, path, query)
```

```python
d = {'file':file, 'id':id, 'method':method, 'path':path, 'query':query, 'url':url, 'label':label}
df = DataFrame(d)
```
